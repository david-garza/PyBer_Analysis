```python
# Import dependencies.
import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create the plot
plt.plot(x_axis,y_axis)
```

```python
# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.plot(x_axis,y_axis)
```

```python
# Create the plot with ax.plt()
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
ax = plt.axes()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot.
plt.plot(x_axis,y_axis)
plt.show()
```

```python
# Create the plot and add a label for the legend.
plt.figure(figsize=(20,10))
plt.plot(x_axis,y_axis, label="Boston", marker="*", color="blue", linewidth=2)
# Create labels for x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45.
plt.ylim(0,45)
#Create a title.
plt.title("PyBer Fare by Month")
# add the legend
plt.legend()
# add grid
plt.grid()
```

```python
#Skill drill object-oriented functions
# Create the figure and set the size
fig, ax = plt.subplots()
ax.plot(x_axis,y_axis, 'D-g', label="Boston", linewidth=2)
ax.legend()
ax.set_title("PyBer Fare by Month")
ax.set_ylabel("Fare ($)")
ax.set_xlabel("Month")
ax.set_ylim(0,45)
ax.grid()
```

```python
# Create a barplot
plt.bar(x_axis,y_axis, color="green", label="Boston")
# Create labels for the x and y axes
plt.xlabel("Month")
plt.ylabel("Fare ($)")
# Create a title
plt.title("PyBer Fare by Month")
# Add the legend
plt.legend()
```

```python
# Skill drill
# 1. Change the bar color to cyan.
# 2. Add a legend for the city of Chicago
# 3. Add a title and axes labels.
# 4. Switch the axis so that the Fare($) data is on the x-axis.
# 5. Invert the y-axis data.
fig,ax = plt.subplots()
ax.barh(x_axis,y_axis, color="cyan",label="Chicago")
ax.set_title("PyBer Fare by Month")
ax.set_xlabel("Fare ($)")
ax.set_ylabel("Month")
ax.invert_yaxis()
ax.legend()
```

```python
# Skill Drill
# Change the color of the markers to sky blue.
# Change the size of the markers to 5 times each data point.
# Make the color 20% transparent.
# Add a black edge color to the circles.
# Make the linewidth of the circles 2 points.
# Add a legend for the city of Boston.
# Add a title.
# Switch the axis so that the Fare($) data is on the x-axis.
# Add a limit for the x-axis data.
# Invert the y-axis so that January is at the top.

y_axis_larger=[]

for data in y_axis:
    y_axis_larger.append(data*5)

fig_, ax = plt.subplots()
ax.scatter(y_axis,x_axis,s=y_axis_larger, c="skyblue",alpha=0.80,edgecolors="black",linewidths=2,label="Boston")
ax.set_title("Fares in Boston by Month")
ax.set_xlabel("Fares ($)")
ax.set_ylabel("Month")
ax.invert_yaxis()
ax.set_xlim(0,50)
```
