```python
%matplotlib inline
```

```python
# Import dependencies.
import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create the plot
plt.plot(x_axis,y_axis)
```

```python
# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.plot(x_axis,y_axis)
```

```python
# Create the plot with ax.plt()
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
ax = plt.axes()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot.
plt.plot(x_axis,y_axis)
plt.show()
```

```python
# Create the plot and add a label for the legend.
plt.figure(figsize=(20,10))
plt.plot(x_axis,y_axis, label="Boston", marker="*", color="blue", linewidth=2)
# Create labels for x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45.
plt.ylim(0,45)
#Create a title.
plt.title("PyBer Fare by Month")
# add the legend
plt.legend()
# add grid
plt.grid()
```

```python
#Skill drill object-oriented functions
# Create the figure and set the size
fig, ax = plt.subplots()
ax.plot(x_axis,y_axis, 'D-g', label="Boston", linewidth=2)
ax.legend()
ax.set_title("PyBer Fare by Month")
ax.set_ylabel("Fare ($)")
ax.set_xlabel("Month")
ax.set_ylim(0,45)
ax.grid()
```

```python
# Create a barplot
plt.bar(x_axis,y_axis, color="green", label="Boston")
# Create labels for the x and y axes
plt.xlabel("Month")
plt.ylabel("Fare ($)")
# Create a title
plt.title("PyBer Fare by Month")
# Add the legend
plt.legend()
```

```python
# Create the plot
plt.barh(x_axis,y_axis)
```

```python
# Create flipped bar chart
plt.barh(y_axis,x_axis)
```

```python
# Invert the y-axis to change the order of the months
plt.barh(x_axis,y_axis)
plt.gca().invert_yaxis()
```

```python
#Bar chart skill drill
plt.barh(x_axis,y_axis, color="magenta", label="Boston")
plt.gca().invert_yaxis()
plt.legend()
plt.xlabel("Fare ($)")
plt.ylabel("Month")
```

```python
# Object-oriented interface method to create a bar chart with skill drill
fig, ax = plt.subplots()
ax.bar(x_axis, y_axis, color="k",label="New York")
ax.legend()
```

```python
# Create a horizontal bar chart using OO
fig,ax = plt.subplots()
ax.barh(x_axis,y_axis)
```

```python
# Skill drill
# 1. Change the bar color to cyan.
# 2. Add a legend for the city of Chicago
# 3. Add a title and axes labels.
# 4. Switch the axis so that the Fare($) data is on the x-axis.
# 5. Invert the y-axis data.
fig,ax = plt.subplots()
ax.barh(x_axis,y_axis, color="cyan",label="Chicago")
ax.set_title("PyBer Fare by Month")
ax.set_xlabel("Fare ($)")
ax.set_ylabel("Month")
ax.invert_yaxis()
ax.legend()
```
